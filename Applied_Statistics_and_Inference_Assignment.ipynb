Ques1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

Ans. In hypothesis testing, we start with a null hypothesis (H₀), which usually states that there is no effect or no difference. We then test data against it, and two types of errors can occur:
- Type I Error (False Positive): This happens when we reject the null hypothesis even though it is actually true. In other words, we conclude that there is an effect when in reality there isn’t. The probability of making this mistake is called alpha (α), often set at 0.05. For example, approving a drug that doesn’t actually work would be a Type I error.
- Type II Error (False Negative): This occurs when we fail to reject the null hypothesis even though it is false. Essentially, we miss a real effect or difference. The probability of this mistake is called beta (β). For instance, failing to approve a drug that truly helps patients would be a Type II error.
These errors directly influence decision-making. A Type I error can lead to wasted resources, false confidence, or harmful consequences if a wrong conclusion is acted upon. A Type II error, on the other hand, can mean missed opportunities, delayed interventions, or overlooking important discoveries.

There’s always a trade-off: lowering the chance of Type I errors (by setting a stricter significance level) often increases the risk of Type II errors, and vice versa. That’s why researchers consider the context—sometimes avoiding false positives is more important (like in legal trials), while in other cases avoiding false negatives matters more (like in medical screenings).

Ques2. What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

Ans. The p-value is the probability of observing your data—or something more extreme—if the null hypothesis were true. A small p-value suggests strong evidence against the null hypothesis, while a large p-value indicates the data are consistent with it.

 What the P-value Represents
- The null hypothesis (H₀) usually states that there is no effect or no difference.
- The p-value quantifies how surprising your observed results would be if H₀ were correct.
- For example, a p-value of 0.03 means there is a 3% chance of seeing results as extreme as yours under the assumption that the null hypothesis is true.

 How to Interpret It
- Small p-value (≤ α, often 0.05): Evidence against H₀ is strong. You reject the null hypothesis and tentatively accept the alternative hypothesis.
- Large p-value (> α): Evidence against H₀ is weak. You fail to reject the null hypothesis, meaning the data are consistent with “no effect.”
It’s important to note:
- A p-value does not tell you the probability that H₀ is true.
- It also does not measure the size or importance of an effect—it only measures consistency with H₀.

 Decision-Making Context
- In medical research, a small p-value might justify approving a new treatment, but researchers must also consider effect size and clinical relevance.
- In quality control, a small p-value could signal a defect in production, prompting corrective action.
- In education or social sciences, researchers use p-values to decide whether observed differences (like test scores between groups) are statistically significant or just due to chance.


**The p-value is a tool for assessing evidence against the null hypothesis, not a definitive truth meter. It should always be interpreted alongside context, effect size, and study design to avoid misleading conclusions.**

Ques3. Explain the difference between a Z-test and a T-test, including when to use each.

Ans. A Z-test is used when the population variance is known and the sample size is large, while a T-test is used when the population variance is unknown and the sample size is small.

 Z-test
- The Z-test relies on the standard normal distribution.
- It is appropriate when the population standard deviation (σ) is known and the sample size is large (n ≥ 30).
- Common uses include comparing a sample mean to a population mean, or comparing two large-sample means when variances are known.
- Example: Testing whether the average height of a large group of students differs from the national average, assuming you know the population variance.

T-test
- The T-test relies on the Student’s t-distribution, which has heavier tails than the normal distribution.
- It is used when the population standard deviation is unknown and must be estimated from the sample.
- Best suited for small sample sizes (n < 30), though it can be used for larger samples too.
- Types include one-sample T-test, independent two-sample T-test, and paired T-test.
- Example: Comparing the average exam scores of two small classes when the population variance is not known.



Practical Decision-Making
- If you have a large dataset and know the population variance, use a Z-test.
- If you have a small dataset or don’t know the variance, use a T-test.
- In practice, researchers often default to the T-test because population variance is rarely known, and modern software makes it easy to compute.

Ques4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?

Ans. A confidence interval is a range of values that estimates a population parameter, and the margin of error determines how wide that range is—the larger the margin of error, the wider and less precise the interval.

A confidence interval (CI) provides a range within which we expect the true population parameter (like a mean or proportion) to lie, based on sample data.
For example, if the average score in a sample is 75 and the margin of error is 3, then the 95% confidence interval is [72, 78]. This means we are 95% confident that the true population mean lies between 72 and 78.

Role of the Margin of Error
The margin of error (MOE) is the amount added and subtracted from the sample estimate to create the interval. It depends on:
- Sample size: Larger samples reduce the margin of error, making the interval narrower and more precise.
- Variability in data: More variability increases the margin of error, widening the interval.
- Confidence level: Higher confidence (e.g., 99% vs. 95%) requires a larger margin of error, widening the interval to ensure greater certainty.
Thus, the margin of error directly controls the width of the confidence interval:
- Small MOE → Narrow interval → More precise estimate
- Large MOE → Wide interval → Less precise estimate

Interpretation in Decision-Making
- A narrow confidence interval suggests strong precision in estimating the population parameter, useful in contexts like medical trials or quality control.
- A wide confidence interval indicates more uncertainty, often due to small sample sizes or high variability, which may limit the reliability of conclusions.
- Importantly, the confidence interval does not guarantee the parameter lies within the range—it reflects the probability based on repeated sampling.


Ques5. Describe the purpose and assumptions of an ANOVA test. How does it
extend hypothesis testing to more than two groups?

Ans. ANOVA (Analysis of Variance) is a statistical method used to test whether the means of three or more groups are significantly different, by comparing the variation between groups to the variation within groups. It extends hypothesis testing beyond two groups, where multiple t-tests would otherwise inflate error rates.

Purpose of ANOVA
The main goal of ANOVA is to determine if observed differences among group means are due to real effects or just random chance.
- With two groups, a t-test suffices.
- With three or more groups, running multiple t-tests increases the risk of Type I error (false positives). ANOVA solves this by testing all groups simultaneously under one framework.
- Example: Comparing average exam scores across three different schools to see if teaching methods differ significantly.

Assumptions of ANOVA
For ANOVA results to be valid, certain assumptions must hold:
- Independence of observations – Each data point is collected independently.
- Normality – The data in each group should be approximately normally distributed.
- Homogeneity of variances – All groups should have similar variance (spread of scores).
- Random sampling – Samples should represent the populations fairly.
Violating these assumptions can lead to misleading conclusions, though robust versions of ANOVA exist to handle deviations.

ANOVA Extends Hypothesis Testing by following ways:
- Null hypothesis (H₀): All group means are equal.
- Alternative hypothesis (H₁): At least one group mean differs.
- ANOVA uses the F-statistic, which is the ratio of variance between groups to variance within groups. A large F-value suggests that differences between group means are unlikely to be due to chance.

Practical Impact
- In education, ANOVA can compare test scores across multiple teaching methods.
- In chemistry experiments, it can test whether different catalysts produce significantly different yields.
- In business, it can compare customer satisfaction across several product lines.

Ques6. Write a Python program to perform a one-sample Z-test and interpret the
result for a given dataset?

Ans. Code written below for a one-sample Z-test checks whether the mean of a sample is significantly different from a known population mean, assuming the population standard deviation is known.


Interpretation of Results
- Z-statistic: Measures how many standard errors the sample mean is away from the hypothesized mean.
- P-value: Probability of observing such a result if the null hypothesis is true.
- Decision Rule:
- If p-value < α (commonly 0.05), reject the null hypothesis.
- Otherwise, fail to reject it.
For example, if the sample mean is 83.2, Z-statistic ≈ 2.5, and p-value ≈ 0.012, we would reject H₀ and conclude that the sample mean is significantly different from 80














```python
# Ques6. Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset?

import numpy as np
from scipy.stats import norm

# Example dataset: exam scores of 15 students
data = np.array([78, 85, 82, 90, 76, 88, 84, 79, 91, 77, 83, 86, 80, 82, 89])

# Known population mean (mu0) and population standard deviation (sigma)
mu0 = 80   # hypothesized population mean
sigma = 5  # assumed known population standard deviation

# Step 1: Calculate sample mean
sample_mean = np.mean(data)

# Step 2: Calculate Z statistic
n = len(data)
z_stat = (sample_mean - mu0) / (sigma / np.sqrt(n))

# Step 3: Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Output results
print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from", mu0)
else:
    print("Fail to reject the null hypothesis: No significant difference from", mu0)
```

```python
# Ques9. Write a Python function to calculate and visualize the confidence interval for a sample mean.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for a sample mean.

    Parameters:
        data (array-like): Sample data
        confidence (float): Confidence level (default 0.95)

    Returns:
        (mean, lower_bound, upper_bound)
    """
    # Convert to numpy array
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean

    # t critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # margin of error
    moe = t_crit * sem

    # confidence interval
    lower_bound = mean - moe
    upper_bound = mean + moe

    # Print results
    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")

    # Visualization
    plt.figure(figsize=(8,4))
    plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(mean, color='red', linestyle='--', label=f"Mean = {mean:.2f}")
    plt.axvline(lower_bound, color='green', linestyle='--', label=f"Lower CI = {lower_bound:.2f}")
    plt.axvline(upper_bound, color='green', linestyle='--', label=f"Upper CI = {upper_bound:.2f}")
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    return mean, lower_bound, upper_bound

# Example usage:
data = np.random.normal(loc=50, scale=10, size=100)  # simulate sample data
confidence_interval(data, confidence=0.95)
```

```python
# Ques10. Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare

# Example: Observed frequencies (say, outcomes of a die rolled 60 times)
observed = np.array([8, 10, 9, 12, 11, 10])

# Expected frequencies (fair die → each outcome equally likely)
expected = np.array([60/6] * 6)  # each face expected 10 times

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Observed distribution differs significantly from expected.")
else:
    print("Fail to reject the null hypothesis: No significant difference between observed and expected.")
```
