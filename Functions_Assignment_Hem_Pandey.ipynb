1. What is the difference between a function and a method in Python?

**Ans.** **A function** in Python is simply a block of reusable code that performs a specific task. It exists on its own, outside of any class, and you call it directly by its name, passing in whatever arguments it needs. It doesn’t automatically know anything about objects or their data unless you explicitly pass that data in.

**A method**, on the other hand, is also a function — but one that lives inside a class and is tied to the objects created from that class. When you call a method on an object, Python automatically passes that object itself as the first argument (usually named self), so the method can work with the object’s own attributes and state.
In short:
- A function is independent and works only with what is given to it.
- A method is bound to an object and can directly use or modify that object’s data.

4. What is the purpose of the `return` statement in a Python function?

**Ans.**
The return statement in Python is what allows a function to send a result back to the place where it was called.
When Python executes return inside a function:
- It immediately stops running the rest of that function’s code.
- It hands back the value (or values) you specify to the caller.

5. What are iterators in Python and how do they differ from iterables?

**Ans.** An iterable is something that can produce an iterator when call iter() on it.
- Examples: lists, tuples, strings, dictionaries, sets, file objects.
- can loop over it multiple times because can always create a fresh iterator from it.

my_list = [10, 20, 30]  # iterable
iterator_obj = iter(my_list)  # creates an iterator from the iterable

An iterator is the object that actually hands the next value when you ask for it.
It remembers its position so it knows where it left off.
- Examples: the object returned by iter(my_list), file readers, generator objects.
- Once it’s exhausted, it can’t be reused — you need a new iterator to loop again.

my_list = [10, 20, 30]
iterator_obj = iter(my_list)  # iterator
print(next(iterator_obj))  # 10
print(next(iterator_obj))  # 20

6. Explain the concept of generators in Python and how they are defined.?

**Ans.**
A generator is a special kind of iterator in Python that produces values on the fly instead of storing them all in memory at once.
- Created using a function with yield or a generator expression
- Automatically handles __iter__() and __next__()
- More concise and memory-efficient


**features:**
- They’re lazy: they only compute the next value when you ask for it.
- They’re stateful: they remember where they left off between calls.
- They’re iterators by nature

**Define a Generator**
- Using a Function with yield: Instead of return, use yield to produce a value and pause the function’s state until the next request.
def density(mass,vol):
  for m in mass:
    yield m/vol
mass=[111,222,333]
d=density(mass,100)
for i in d:
  print(i)

Each yield sends one density value to the caller.

- Generator Expressions: Like list comprehensions, but with parentheses — they create a generator instead of a list.
Mass_data = [2400, 2500, 2300]
density = (mass /1000 for mass in mass_data)

for d in density:
    print(d)

7. What are the advantages of using generators over regular functions?

**Ans.**
**Memory Efficiency**
- Regular function: Builds the entire result in memory before returning it.
- Generator: Produces one value at a time, on demand, without storing the whole dataset.
- Benefit: can process huge datasets without running out of RAM
**Faster Startup Time**
- Generators start producing results immediately — no wait for the whole computation to finish before we see the first value.
- This is perfect for streaming dashboards or real-time analytics.

**Lazy Evaluation**
- Values are computed only when needed.
- saving time and CPU cycles.

**Infinite Sequences**
- Generators can represent endless data streams without ever building a “full” list.
- Regular functions can’t return infinite lists — they’d run forever or crash

8. What is a lambda function in Python and when is it typically used?

**Ans.** A lambda function in Python is a small, anonymous function defined with the lambda keyword instead of def.
It can take any number of arguments but can only contain a single expression — that expression is automatically returned.

They’re typically used when:
- need a quick, throwaway function — one you don’t plan to reuse elsewhere.
- want to pass a function as an argument to another function (e.g., sorting, filtering, mapping).
- want concise code for simple operations.

eg with map()
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)  # [1, 4, 9, 16]

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

for num in fibonacci(int(input("enter num of terms of fibonacci: "))):
    print(num, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

from docx import Document

def read_docx_lines(filename):
    doc = Document(filename)
    for para in doc.paragraphs:
        yield para.text.strip()

for line in read_docx_lines("/content/What is PW Skills.docx"):
    print(line)

```
