Ques1. What is Simple Linear Regression (SLR)? Explain its purpose.?

Ans. Simple Linear Regression (SLR) is a statistical method used to model the relationship between two variables:
- Independent variable (X): The predictor or input.
- Dependent variable (Y): The outcome or response.
The relationship is expressed using a straight-line equation:
Y=\c +bX+e
- c: Intercept (value of Y when X = 0)
- b: Slope (how much Y changes for a one-unit change in X)
- e: Error term (captures variability not explained by X)

The main goals of Simple Linear Regression are:
- Prediction: Estimate the value of the dependent variable (Y) based on a given independent variable (X).
Example: Predicting solar power output based on sunlight hours.
- Understanding Relationships: Quantify how strongly and in what direction X influences Y.
Example: Measuring how temperature affects electricity consumption.
- Trend Identification: Detect linear patterns in data to simplify complex relationships.
Example: Identifying whether sales increase steadily with advertising spend.
- Decision-Making: Provide evidence-based insights for planning, forecasting, and optimization.
Example: In solar PV design, regression can help forecast generation based on irradiance levels.

Ques5. What is the method of least squares in linear regression?

Ans. Method of Least Squares in Linear Regression:

The method of least squares is the standard technique used to estimate the coefficients of a linear regression model. Its purpose is to find the line that best fits the data by minimizing the overall error between observed and predicted values.

- Regression Equation:
Y=c+bX+e
- c: Intercept
- b: Slope
- e : Error term


Residuals (Errors):
For each data point, the residual is:
e=Y-Y'
- where y'=c+bX.

Objective Function (Sum of Squared Errors):
The least squares method minimizes the sum of squared residuals:
SSE=sum(Y_i-(c+bX_i))^2
- Optimization:
- Take partial derivatives of SSE with respect to c and b.
- Set them equal to zero to solve for the values of c and b.

Ques6.   What is Logistic Regression? How does it differ from Linear Regression?

Ans. Logistic Regression is used for classification problems, while Linear Regression is used for prediction of continuous values. The key difference lies in the type of dependent variable and the mathematical function used to model the relationship.

 Logistic Regression
- Definition: A statistical model used to predict the probability of a categorical outcome (often binary: Yes/No, 0/1).
- Equation:
P(Y=1|X)=\frac{1}{1+e^{-(c+bX)}}- Produces probabilities between 0 and 1 using the sigmoid function.
- Purpose: Classification tasks (e.g., spam vs. not spam, disease vs. no disease).
- Estimation Method: Maximum Likelihood Estimation (MLE).

 Linear Regression
- Definition: A statistical model used to predict continuous numerical outcomes.
- Equation:
Y=c+bX+epsilon
- Produces values across the real number line.
- Purpose: Regression tasks (e.g., predicting house prices, solar PV output).
- Estimation Method: Least Squares (minimizing squared errors).

Ques7. Name and briefly describe three common evaluation metrics for regression
models.?

Ans. ðŸ“Š Three Common Evaluation Metrics for Regression Models
When evaluating regression models, we want to measure how well the predicted values match the actual values. Here are three widely used metrics:

- Mean Absolute Error (MAE)

- Measures the average absolute difference between predicted and actual values.
- Easy to interpret: it tells us, on average, how far off predictions are.
- Example: If MAE = 5, predictions are off by 5 units on average.

- Mean Squared Error (MSE)

- Squares the errors before averaging, penalizing larger errors more heavily.
- Useful when large deviations are particularly undesirable.
- Example: In solar PV forecasting, a big error on peak generation days would be penalized strongly.

- R-squared (R^2) â€“ Coefficient of Determination

- Represents the proportion of variance in the dependent variable explained by the model.
- Ranges from 0 to 1 (closer to 1 means better fit).
- Example: An R^2 of 0.85 means 85% of the variation in PV output is explained by irradiance.

Ques8. What is the purpose of the R-squared metric in regression analysis?

Ans. R-squared (R^2), also called the coefficient of determination, is a statistical measure that explains how well a regression model fits the data.


R^2=1-Sum  of Squared  Errors (SSE)/Total  Sum  of  Squares(SST)
- SSE (Error): Variation in Y not explained by the model.
- SST (Total): Total variation in Y around its mean.
- RÂ² Value: The proportion of variance in the dependent variable explained by the independent variable(s).

Purpose
- Goodness of Fit: Shows how well the regression line represents the data.
- Interpretability:
- R^2=0: Model explains none of the variability.
- R^2=1: Model explains all variability perfectly.
- Values between 0 and 1 indicate partial explanation.
- Comparison Tool: Helps compare different regression models to see which fits better.


Ques9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.



```python
#Ans9.
# Import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Example dataset
# X = independent variable (e.g., solar irradiance)
# Y = dependent variable (e.g., PV output)
X = np.array([100, 200, 300, 400, 500]).reshape(-1, 1)  # must be 2D
Y = np.array([50, 100, 150, 200, 250])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print slope (coefficient) and intercept
print("Slope (b):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```
