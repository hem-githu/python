Ques1. What is the difference between AI, ML, DL, and Data Science? Provide a
brief explanation of each.?

Ans.
- AI (Artificial Intelligence): Broad field of making machines mimic human intelligence.
- ML (Machine Learning): Subset of AI where systems learn patterns from data.
- DL (Deep Learning): Subset of ML using neural networks with many layers for complex tasks.
- Data Science: Discipline focused on extracting insights from data using statistics, ML, and domain knowledge.

Artificial Intelligence (AI)
- Definition: AI is the science of creating systems that can perform tasks requiring human-like intelligence, such as reasoning, problem-solving, and decision-making.
- Scope: Very broad — includes rule-based systems, expert systems, robotics, natural language processing, and machine learning.
- Example: A chatbot that understands and responds to customer queries.

 Machine Learning (ML)
- Definition: ML is a subset of AI that enables systems to learn automatically from data and improve performance without explicit programming.
- Scope: Focuses on algorithms that detect patterns and make predictions.
- Example: Predicting house prices based on features like area and number of rooms.

 Deep Learning (DL)
- Definition: DL is a specialized branch of ML that uses artificial neural networks with multiple layers (deep architectures).
- Scope: Excels at handling large, complex datasets like images, audio, and text.
- Example: Image recognition (e.g., identifying cats vs. dogs in photos).


Data Science
- Definition: Data Science is a multidisciplinary field that combines statistics, programming, and domain expertise to extract insights and knowledge from data.
- Scope: Broader than AI/ML — includes data cleaning, visualization, hypothesis testing, and storytelling with data.
- Example: Analyzing customer purchase behavior to recommend marketing strategies

Ques2. : Explain overfitting and underfitting in ML. How can you detect and prevent them?

Ans.
- Overfitting happens when a model learns the training data too closely (including noise), performing well on training but poorly on unseen data.
- Underfitting happens when a model is too simple, failing to capture important patterns, leading to poor performance on both training and test data.
- You can detect them by comparing training vs. validation/test performance, and prevent them using techniques like regularization, cross-validation, and proper feature selection

Overfitting:

 Definition: The model memorizes training data, including noise and outliers, instead of learning general patterns.

Symptoms:
- Very high accuracy on training data.
- Poor accuracy on validation/test data.
- Residuals show systematic patterns.

Causes:
- Too complex model (many parameters, deep trees, high-degree polynomials).
- Insufficient training data.
- Training too long without regularization.

Prevention:
- Use regularization (L1/Lasso, L2/Ridge).
- Apply dropout in neural networks.
- Use cross-validation to tune hyperparameters.

Underfitting:

 Definition: The model is too simple and fails to capture the underlying trend.

 Symptoms:
- Low accuracy on both training and test data.
- Predictions are biased and oversimplified.

Causes:
- Model lacks complexity (e.g., linear regression for non-linear data).
- Too few features or poor feature engineering.
- Insufficient training (early stopping too soon).

Prevention:
- Use a more complex model (e.g., polynomial regression, deeper trees).
- Add relevant features or improve feature engineering.
- Train longer with proper tuning.


How to Detect
- Train vs. Validation/Test Performance:
- Overfitting → High train accuracy, low test accuracy.
- Underfitting → Low train accuracy, low test accuracy.
- Learning Curves:
- Plot error vs. training size. Divergence indicates overfitting; consistently high error indicates underfitting.
- Cross-validation: Confirms whether the model generalizes well

Ques3. How would you handle missing values in a dataset? Explain at least three
methods with examples?

Ans.
Missing values can be handled by deletion, imputation (mean/median/mode), or advanced techniques like regression or KNN imputation. The choice depends on the dataset size, type of variable, and importance of the missing data.
1. Deletion Methods
- Listwise Deletion: Remove entire rows with missing values.
- Example: If a dataset of house prices has missing values for "Rooms" in 5 rows, those rows are dropped.
- Pros: Simple, keeps dataset clean.
- Cons: Risk of losing valuable data, especially if missingness is large.
- Column Deletion: Drop a feature if it has too many missing values (e.g., >50%).
- Example: If "Garage Size" is missing in 70% of entries, drop the column.

 2. Imputation with Simple Statistics
- Mean/Median Imputation: Replace missing values with the mean or median of that column.
- Example: In a dataset of student scores, if "Math Score" is missing, fill it with the average score.
- Pros: Easy to implement, works well for numerical data.
- Cons: Can distort variance and correlations.
- Mode Imputation: Replace missing categorical values with the most frequent category.
3. Advanced Imputation Techniques
- Regression Imputation: Predict missing values using other features.
- Example: If "House Price" is missing, predict it using "Area" and "Rooms" via regression.
- Pros: More accurate than simple imputation.
- Cons: Computationally expensive, assumes linear relationships.
- KNN Imputation: Use the values from the nearest neighbors to fill missing data.
- Example: If "Income" is missing, find the 5 nearest neighbors based on age and education, then average their incomes.
- Pros: Captures complex relationships.
- Cons: Slower for large datasets.

Ques4. What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).

Ans. An imbalanced dataset occurs when the classes in a classification problem are not represented equally. One class has many more samples than the other(s).
- Example: In fraud detection, 99% of transactions are legitimate and only 1% are fraudulent.
- Problem: Standard machine learning models tend to favor the majority class, leading to poor performance on the minority class (the one we often care about most).

Technique 1: Resampling Methods

Theoretical Explanation

Resampling adjusts the dataset to balance class distribution:
- Oversampling: Duplicate or synthetically generate minority class samples.
- Undersampling: Reduce majority class samples to match minority class size.

Technique 2: Algorithmic Approaches

Theoretical Explanation
Instead of changing the dataset, modify the learning algorithm:
- Class Weighting: Assign higher penalty to misclassifying minority class.
- Cost-Sensitive Learning: Model learns to prioritize minority class errors.


Ques5. Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.?

Ans.Why Feature Scaling is Important in ML
Feature scaling ensures that all input variables contribute proportionally to the model. Many machine learning algorithms are distance-based (like KNN, SVM, clustering) or gradient-based (like logistic regression, neural networks). If features are on different scales, the model may:
- Give more weight to variables with larger ranges (e.g., income in thousands vs. age in years).
- Struggle with convergence in optimization (gradient descent takes longer).
- Produce distorted distance measures (Euclidean distance dominated by one feature)

Min-Max Scaling (Normalization):


- Use Case: Useful when you know the data has bounded values (e.g., pixel intensities between 0–255).
- Pros: Preserves relationships and distribution shape.
- Cons: Sensitive to outliers — one extreme value can shrink the scale of all other data.
Example:
- Age = [20, 40, 60] → Min-Max scaling → [0, 0.5, 1]

 Standardization (Z-score Normalization)

- Use Case: Preferred when data has outliers or unbounded ranges. Common in algorithms assuming Gaussian distribution (e.g., logistic regression, PCA).
- Pros: Less sensitive to outliers compared to Min-Max.
- Cons: Doesn’t keep values in a fixed range.
Example:
- Age = [20, 40, 60], Mean = 40, Std = 20 → Standardization → [-1, 0, +1]

Ques6. Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
Hint: Consider categorical variables with ordinal vs. nominal relationships

Ans. Label Encoding and One-Hot Encoding are two common techniques for converting categorical variables into numerical form, but they serve different purposes depending on the type of categorical data
Label Encoding
- Definition: Assigns each category a unique integer value.

- Best For: Ordinal variables (categories with a natural order).
- Example: Education level (High School < Bachelor < Master < PhD).
- Pros: Simple, compact representation.
- Cons: Implies an order even if none exists (e.g., encoding "Red=0, Blue=1, Green=2" suggests Red < Blue < Green).

One-Hot Encoding
- Definition: Creates a new binary column for each category, marking presence with 1 and absence with 0.
- Best For: Nominal variables (categories with no inherent order).
- Example: City names, product types, colors.
- Pros: Avoids false ordering, preserves categorical meaning.
- Cons: Increases dimensionality (many new columns if categories are numerous).


Ques7. Google Play Store Dataset a). Analyze the relationship between app categories and ratings. Which categories have the highest/lowest average ratings, and what could be the possible reasons?
Dataset: https://github.com/MasteriNeuron/datasets.git

Ans. Risks & Limitations
- Bias in Ratings: Some categories attract more engaged users who rate positively.
- Data Skew: Popular apps dominate averages; niche apps may distort category means.
- Temporal Effects: Ratings can change over time with updates or new competitors.

Key Takeaway
- Educational and wellness apps consistently earn higher ratings due to reliability and usefulness.
- Dating, finance, and utility apps struggle with user trust, monetization, and quality consistency, leading to lower ratings.



```python
#ques7 code
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/refs/heads/main/googleplaystore.csv"
data = pd.read_csv(url)

print(data.head())



# Clean data: drop rows with missing or invalid ratings
data = data[pd.to_numeric(data['Rating'], errors='coerce').notnull()]
data['Rating'] = data['Rating'].astype(float)

# Group by category and compute average rating
category_ratings = data.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Print results
print("Average Ratings by Category:")
print(category_ratings)

# Plot bar chart
plt.figure(figsize=(12,6))
category_ratings.plot(kind='bar', color='skyblue')
plt.title("Average App Ratings by Category (Google Play Store)")
plt.ylabel("Average Rating")
plt.xlabel("Category")
plt.xticks(rotation=90)
plt.show()
```

```python
#Ques8. Titanic Dataset a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?
#b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival?
# code
import pandas as pd

# Load Titanic dataset (assuming titanic.csv is in your working directory)
url= "https://raw.githubusercontent.com/MasteriNeuron/datasets/refs/heads/main/titanic.csv"
data = pd.read_csv(url)

print(data.head())
# --- a) Survival rates by passenger class ---
pclass_survival = data.groupby('Pclass')['Survived'].mean()

print("Survival Rates by Passenger Class:")
print(pclass_survival)

# --- b) Survival rates by age groups (children vs adults) ---
# Create age group column
data['AgeGroup'] = data['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

agegroup_survival = data.groupby('AgeGroup')['Survived'].mean()

print("\nSurvival Rates by Age Group:")
print(agegroup_survival)

# --- Optional: Visualization ---
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))

# Plot survival by class
plt.subplot(1,2,1)
pclass_survival.plot(kind='bar', color='skyblue')
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.xlabel("Passenger Class")

# Plot survival by age group
plt.subplot(1,2,2)
agegroup_survival.plot(kind='bar', color='lightgreen')
plt.title("Survival Rate by Age Group")
plt.ylabel("Survival Rate")
plt.xlabel("Age Group")

plt.tight_layout()
plt.show()
```
